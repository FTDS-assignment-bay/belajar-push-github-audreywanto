# Caranya:

1. copy ssh `link`
2. **git clone** `ssh link` -> output nya akan membuat folder baru
3. cd ke new folder tersebut
4. **git status** -> outputnya akan muncul nama-nama file di folder baru tsb berwarna merah
5. *git add .* -> artinya saya akan ngepush seluruh file di folder baru tsb

bisa juga `git add push_kuy.ipynb` -> untuk ngepush file push_kuy saja

6. **git status** -> outputnya akan muncul nama-nama file di folder baru yg sudah di add berwarna hijau
7. **git commit** -m "revisi_1"
8. `git push`
9. Jangan lupa untuk refresh repo kalian, untuk make sure file sudah terpush/terupdate
